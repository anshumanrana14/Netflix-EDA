```python
import pandas as pd
netflix = pd.read_csv('netflix_titles.csv')
netflix.head()
```

```python
netflix.describe()
```

```python
netflix.isnull().sum()
```

```python
exact_duplicates = netflix.duplicated().sum()
exact_duplicates
```

```python
netflix=netflix.dropna(subset=['duration','rating'])
netflix
```

```python
netflix = netflix[netflix['type'].isin(['Movie', 'TV Show'])].copy()
netflix['country'] = netflix['country'].fillna('Unknown')
netflix['director'] = netflix['director'].fillna('Not Available')
netflix['cast'] = netflix['cast'].fillna('Not Listed')
netflix['date_added'] = netflix['date_added'].fillna('Date Not Available')
netflix.isnull().sum()
```

```python
netflix.head(10)
```

```python
netflix['type'].value_counts()
```

```python
top_directors=netflix['director'].value_counts().head(8)
top_directors
```

```python
top_country=netflix['country'].value_counts().head(15)
top_country
```

```python
top_release_year=netflix['release_year'].value_counts().head(30)
top_release_year
```

```python
top_ratings=netflix['rating'].value_counts().head(10)
top_ratings
```

```python
Top_types_by_country=netflix.groupby(['country','type']).size().reset_index(name='count')
Top_types_by_country = Top_types_by_country.sort_values(by='count', ascending=False).head(20)
Top_types_by_country
```

```python
directors_summary=netflix.groupby(['director','type','listed_in']).size().reset_index(name='Counts')
directors_summary=directors_summary.sort_values(by='Counts',ascending=False).head(20)
directors_summary
```

```python
netflix['duration_min'] = netflix['duration'].str.extract(r'(\d+)').astype(float)
netflix
```

```python
documentaries=netflix[netflix['listed_in'].str.contains('Documentaries')]
documentaries
Selected_columns=documentaries[['title','type','listed_in','description','duration_min']]
Selected_columns.head(20)
```

```python
top_ratings = netflix['rating'].value_counts().head(5).index.tolist()
top_rating_shows = netflix[netflix['rating'].isin(top_ratings)]
top_rating_shows = top_rating_shows[['title', 'rating', 'release_year', 'country', 'director']]
top_rating_shows.head(10)
```

```python
import matplotlib.pyplot as plt
type_counts = netflix['type'].value_counts()# this is used to count the content
type_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Count of Movies vs TV Shows on Netflix')
plt.tight_layout()
plt.show()
```

```python
top_directors=netflix['director'].value_counts().head()
plt.figure(figsize=(5,6))
top_directors.plot(kind="bar",color=["blue"])
plt.xlabel("Names of the directors")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt


top_ratings = netflix['rating'].value_counts().head(10)

# Plot bar chart
plt.figure(figsize=(10, 8))
top_ratings.plot(kind="bar", color="skyblue")
plt.title("Content Distribution by Rating")
plt.xlabel("Rating")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

countries = ['India', 'Japan', 'United States', 'United Kingdom', 'Unknown']
data = netflix[netflix['country'].isin(countries)]
data.groupby(['country', 'type']).size().plot(
    kind='bar',
    figsize=(10, 6),
    color=['Black', 'gold']
)
plt.title("Movies vs TV Shows by Country")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```
